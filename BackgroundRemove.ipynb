```python
# background_remover.py
# Usage in Jupyter: from background_remover import remove_background; remove_background("photo.jpg")

import os, sys, subprocess

# Ensure rembg is installed
try:
    from rembg import remove
except Exception:
    subprocess.check_call([sys.executable, "-m", "pip", "install", "rembg", "--quiet"])
    from rembg import remove

def remove_background(input_path, output_path=None, alpha_matting=True,
                      alpha_matting_foreground_threshold=240,
                      alpha_matting_background_threshold=10):
    """Remove background from input image and save PNG with transparency.
    Returns the output file path."""
    with open(input_path, "rb") as f:
        input_bytes = f.read()

    out_bytes = remove(
        input_bytes,
        alpha_matting=alpha_matting,
        alpha_matting_foreground_threshold=alpha_matting_foreground_threshold,
        alpha_matting_background_threshold=alpha_matting_background_threshold,
    )

    if output_path is None:
        base, _ = os.path.splitext(input_path)
        output_path = base + "_no_bg.png"

    with open(output_path, "wb") as f:
        f.write(out_bytes)

    return output_path

# Optional helper to process a directory
def remove_background_folder(src_folder, dst_folder=None, ext=(".png", ".jpg", ".jpeg")):
    if dst_folder is None:
        dst_folder = os.path.join(src_folder, "no_bg")
    os.makedirs(dst_folder, exist_ok=True)
    results = []
    for name in os.listdir(src_folder):
        if name.lower().endswith(ext):
            src = os.path.join(src_folder, name)
            dst = os.path.join(dst_folder, os.path.splitext(name)[0] + "_no_bg.png")
            remove_background(src, dst)
            results.append(dst)
    return results
```
